```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("/content/orders.csv")
```

```python
users = pd.read_json("/content/users.json")
```

```python
conn = sqlite3.connect("/content/restaurants.db")

with open("/content/restaurants.sql", "r") as file:
    sql_script = file.read()

conn.executescript(sql_script)

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
```

```python
orders_users = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)
```

```python
final_df = pd.merge(
    orders_users,
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
if "restaurant_name_y" in final_df.columns:
    final_df.rename(columns={"restaurant_name_y": "restaurant_name"}, inplace=True)

if "restaurant_name_x" in final_df.columns:
    final_df.drop(columns=["restaurant_name_x"], inplace=True)
```

```python
final_df.to_csv("/content/final_food_delivery_dataset.csv", index=False)
```
